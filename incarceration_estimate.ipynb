```python
import census_explorer as ce
import pandas as pd
```

```python
# function which gets demographic tract data for each state. 
def get_tract_data(state):
    vintage = {"year": 2021, "dataset": "acs", "estimate": "acs5"}
    huge_table = ce.table_fetch(table='B02001', vintage=vintage, geography=['tract', ['state', '09']])
    populations = ce.to_pandas(huge_table, readable_columns=True)
    populations['geoid'] = populations['state'] + populations['county'] + populations['tract']
    return populations
```
